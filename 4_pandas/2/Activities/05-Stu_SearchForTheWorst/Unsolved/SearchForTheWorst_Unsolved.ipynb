```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Create reference to CSV file
filepath = os.path.join('..', 'resources', 'Soccer2018Data.csv')
```

```python
# Import the CSV into a pandas DataFrame
soccer_df = pd.read_csv(filepath)
soccer_df.head(3)
```

```python
# List of all of the unique values within the "Preferred Position" column
positions = soccer_df['Preferred Position'].unique().tolist()
positions
```

```python
# Filter the position to only include Strikers (ST)
# Save this to a new DataFrame called `strikers_2018_df`
strikers_2018_df = soccer_df[soccer_df['Preferred Position'] == 'ST']
strikers_2018_df.head(3)
```

```python
# Sort the DataFrame by the values in the "ST" column
strikers_sorted_df = strikers_2018_df.sort_values('ST')
strikers_sorted_df.head(3)
```

```python
# Reset the index of the sorted DataFrame so that the worst player is at index 0
strikers_sorted_df.reset_index(drop=True ,inplace=True)
strikers_sorted_df.head(3)
```
