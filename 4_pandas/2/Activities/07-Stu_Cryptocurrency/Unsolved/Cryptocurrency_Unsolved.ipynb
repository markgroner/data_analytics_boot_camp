```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Establish file paths 
bitcoin_path = os.path.join('..','Resources','bitcoin_cash_price.csv')
dash_path = os.path.join('..','Resources','dash_price.csv')
```

```python
# Read each csv file into a dataframe
bitcoin_df = pd.read_csv(bitcoin_path)
dash_df = pd.read_csv(dash_path)
display(bitcoin_df.head(3))
display(dash_df.head(3))
```

```python
# Merge the two DataFrames together based on the Dates they share
crypto_df = pd.merge(bitcoin_df, dash_df, on='Date')
crypto_df.head(3)
```

```python
# Rename columns so that they are differentiated
crypto_df = crypto_df.rename(columns={"Open_x":"Bitcoin Open","High_x":"Bitcoin High","Low_x":"Bitcoin Low",
                                      "Close_x":"Bitcoin Close","Volume_x":"Bitcoin Volume","Market Cap_x":"Bitcoin Market Cap"})

crypto_df = crypto_df.rename(columns={"Open_y":"Dash Open","High_y":"Dash High","Low_y":"Dash Low",
                                      "Close_y":"Dash Close","Volume_y":"Dash Volume","Market Cap_y":"Dash Market Cap"})

crypto_df.head()
```

```python
# Calculate the best opening price for Bitcoin and Dash
bitcoin_open_max = crypto_df['Bitcoin Open'].max()
dash_open_max = crypto_df['Dash Open'].max()
print(bitcoin_open_max)
print(dash_open_max)
```

```python
# Calculate the best closing price for Bitcoin and Dash
bitcoin_close_max = crypto_df['Bitcoin Close'].max()
dash_close_max = crypto_df['Dash Close'].max()
print(bitcoin_close_max)
print(dash_close_max)
```

```python
# Calculate the total volume for Bitcoin and Dash
bitcoin_volume_sum = crypto_df['Bitcoin Volume'].max()
dash_volume_sum = crypto_df['Dash Volume'].max()
print(bitcoin_volume_sum)
print(dash_volume_sum)
```

```python
# Creating a summary DataFrame using above values
crypto_summary_df = pd.DataFrame({'Currency':['Bitcoin','Dash'], 'Max Open':[bitcoin_open_max, dash_open_max], 'Max Close':[bitcoin_close_max, dash_close_max], 'Total Volume':[bitcoin_volume_sum, dash_volume_sum]})
crypto_summary_df
```
