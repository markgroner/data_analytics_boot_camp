```python
import requests
import json
```

```python
# Make a request to the following endpoint (http://nyt-mongo-scraper.herokuapp.com/api/headlines)
# and store the response
nyt_time_url = 'http://nyt-mongo-scraper.herokuapp.com/api/headlines'
nyt_response = requests.get(nyt_time_url)
nyt_response
```

```python
# JSON-ify the response
nyt_json = nyt_response.json()
nyt_json
```

```python
# Print the JSON representations of the first post
num_posts = len(nyt_json)
first_json = nyt_json[0]
first_json
```

```python
# Print the JSON representations of the last post
last_json = nyt_json[-1]
last_json
```
