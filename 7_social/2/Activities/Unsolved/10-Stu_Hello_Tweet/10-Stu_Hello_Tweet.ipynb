```python
# Dependencies
import tweepy
import json

# Twitter API Keys
from config import twitter_keys
consumer_key = twitter_keys['consumer_key']
consumer_secret = twitter_keys['consumer_secret']
access_token = twitter_keys['access_token']
access_token_secret = twitter_keys['access_token_secret']

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# Note: Twitter prevents tweeting the same status or message multiple
# times. Be sure to change the text when testing.

# Create two status updates
api.update_status('helllloooooooo')
api.update_status('not a bot... not a bot... I am not a bot')
```

```python
# Create a status update with an image
api.update_with_media('pug_pup.jpg', 'pug puppy')
```

```python
# Create a friendship with another user
api.create_friendship('jiji_v_jacob')
```
