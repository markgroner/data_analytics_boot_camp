```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fahrenheit_array = np.array([39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44])
fahrenheit_array
```

```python
months_array = np.arange(1,13)
months_array
```

```python
plt.plot(months_array, fahrenheit_array)
plt.show()
```

```python
celsius_array = [(temp - 32) * (5/9) for temp in fahrenheit_array]
plt.plot(months_array, celsius_array)
plt.show()
```

```python
plt.plot(months_array, fahrenheit_array)
plt.plot(months_array, celsius_array)
plt.show()
```
