```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(len(gyms))
plt.bar(x_axis,members, align='edge')
# Tell matplotlib where we would like to place each of our x axis headers
tick_locations = [value+0.4 for value in x_axis]
plt.xticks(tick_locations, gyms)
# Give our chart some labels and a tile
plt.title('Total Members by Gym')
plt.xlabel('Gym Name')
plt.ylabel('Number of Members')
plt.show()
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.plot(x_axis, sin)
plt.xlim(0, x_lim)
plt.show()
```

```python
# DATASET 3
planet_fitness_ages = ["18-25", "26-33", "34-41", "42+"]
members = [31, 27, 18, 16]
x_axis = np.arange(0, len(planet_fitness_ages))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0.05, 0, 0, 0)
```

```python
plt.pie(members, explode=explode, labels=planet_fitness_ages, shadow=False,
        colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
# DATASET 4
sq_ft = [929, 1000, 693, 1320, 821, 925, 750]
house_price = [285000, 300000, 250000, 315000, 275000, 280000, 260000]
```

```python
plt.scatter(sq_ft, house_price)
# Create a title, x label, and y label for our chart
plt.title("House Price v Square Foot")
plt.xlabel("Square Ft")
plt.ylabel("House Price (Dollars)")
plt.show()
```
